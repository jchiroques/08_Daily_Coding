```python
from os import scandir

def ls2(path=''):
    return [file.name for file in scandir(path)]

ls2('../datos/gatos/fotos')
```

```python
import cv2 
import matplotlib.pyplot as plt 

# Cargamos el modelo para cada patron de deteccin hay un modelo diferente. 
cat_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalcatface.xml') 

# Cargamos la imagen a leer 
img = cv2.imread('../datos/gatos/fotos/gato_4.jpeg') 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

# detectar caras de gatos 
cats = cat_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=3) 

# Dibujar rectangulos alrededor de las caras 
for (x, y, w, h) in cats: 
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2) 
    
# Mostrar imagen 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.imshow(img_rgb)
```
