```python
# Problema del sheriff

# Librerias
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 40, 1_000) # Inicio, fin, n puntos

d_r = 2.5*t
d_s = 3*(t-5)
```

```python
fig, ax = plt.subplots()

plt.title('Gr√°fico de problema del sheriff')
plt.xlabel('Tiempo')
plt.ylabel('Distancia')

ax.set_xlim([0, 40])
ax.set_ylim([0, 100])

ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')

plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```
